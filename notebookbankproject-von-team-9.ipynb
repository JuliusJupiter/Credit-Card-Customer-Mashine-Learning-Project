```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_trainNorm, y_train)

y_predKnn = knn.predict(X_testNorm)
from sklearn.metrics import ConfusionMatrixDisplay
cmdKnn = ConfusionMatrixDisplay.from_predictions(y_test, y_predKnn)
cmdKnn
```

# DecisionTree

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
y_predTree = clf.predict(X_test)
cmdTree = ConfusionMatrixDisplay.from_predictions(y_test, y_predTree)
cmdTree
```

```python
print(classification_report(y_test, y_predTree))
```

**Wir suchen nun die besten Hyperparameter für den tree Classifier. Dazu verwenden wir GridsearchCV.
Erst haben wir zu viele Hyperparameter verwendet, was dazu geführt hat, dass wir sehr lange warten mussten.
Wir haben dann entschieden die Parameter einzugrenzen. (siehe paramTree variable)
Wir verwenden 5er-Schritte, bzw. 2er-Schritte bei max_depth, um ein ungefähren Bereich zu ermitteln.**

```python
#paramTree =  {"min_samples_split": range(5, 50), "max_depth": range(1,30), "criterion": ["gini", "entropy"], "min_samples_leaf":range(1,50)}
paramTree =  {"min_samples_split": range(2, 15, 2), "max_depth": range(2,14,2), "min_samples_leaf":range(1,15)}
gridclf = GridSearchCV(clf, paramTree, n_jobs = -1, return_train_score = True)
gridclf.fit(X_train, y_train)
gridclf.best_params_
```

**Nun grenzen wir die Hyperparameter noch enger ein. Wir wissen nämlich schon, in welchem Bereich sich die optimalen Parameter befinden und können somit den GridSearch in kleineren Abständen durchaufen lassen, ohne dass der Prozess zu lange dauert.**

```python
paramTree =  {"min_samples_split": range(2, 6), "max_depth": range(4,8), "min_samples_leaf":range(1,10)}
gridclf = GridSearchCV(clf, paramTree, n_jobs = -1, return_train_score = True)
gridclf.fit(X_train, y_train)
gridclf.best_params_
```

**Wir trainieren nun den DecisionTreeClassifier mit den durch die GridSearch herausgefundenen Parameter.**

```python
clf2 = tree.DecisionTreeClassifier(max_depth=5, min_samples_leaf= 1, min_samples_split= 2, random_state=42)
clf2.fit(X_train, y_train)
y_predTree2 = clf2.predict(X_test)
cmdTree2 = ConfusionMatrixDisplay.from_predictions(y_test, y_predTree)
cmdTree2
```

**Mit den erhaltenen Hyperparametern aus dem GridSearch erhalten wir in sämtlichen Kategorien bessre Werte.**

```python
print(classification_report(y_test, y_predTree2))
```

# **Möglicher Geschäftseinsatz**
**Unser Modell ist in der Lage, die Kunden, welche ihre Kreditkarte kündigen, zu erkennen! Die Bank (das Unternehmen) könnte dieses Modell nutzen, um vorherzusagen, welcher ihrer Kunden potenziell als nächstes die Geschäftsbeziehung beenden wird. Für diese Nutzung wäre es sinnvoll, einen hohen Recall-Wert zu erzielen. Dies begründet sich darin, dass die Bank "kipp-Kunden", möglichst schon bevor sie kündigen, erkennen will. Wenn der Recall-Wert erhöht wird, werden auch Customer erkannt, die tendenziell in das Muster eines "Attrited" Customers passen, obwohl sie noch nicht gekündigt haben. Um diese kann die Bank sich dann besonders bemühen. Die Bank möchte nämlich möglichst viele solcher "kipp-Kandidaten" identifizieren.**
